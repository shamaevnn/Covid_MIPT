```python
import numpy as np
import pandas as pd
import os
```

```python
directory = '/Users/botanovaolga/Desktop/mipt/factorial/Covid_MIPT/Russia_region'

for filename in os.listdir(directory):
    filepath = directory + '/' + filename
    try:
        file = pd.read_csv(filepath)
        file['daily_cases_lag'] = file['total_cases'].shift(periods = 1).fillna(0)
        file['daily_death_lag'] = file['total_death'].shift(periods = 1).fillna(0)
        file['daily_cases'] = file['total_cases'] - file['daily_cases_lag']
        file['total_currently_infected'] = file['total_cases'] - file['cured']
        file['death_per_day'] = file['total_death'] - file['daily_death_lag']
        file = file.drop(columns = ['cured', 'daily_cases_lag', 'daily_death_lag'])
        file.to_csv(filepath, index = False)
    except:
        pass
```
